# HUMAN -> bio_to_entities 
(O O O O B-LOC O B-LOC O) -> [{'entity': '..', 'label': '...'}, ...

```python
def bio_to_entities(sentence_text, ner_tags_str):
    tokens = sentence_text.split()
    tags = ner_tags_str.split()

    entities = []
    current_tokens = []
    current_class = None

    for tok, tag in zip(tokens, tags):
        if tag == "O": # end entity
            if current_tokens:
                entities.append({
                    "entity": " ".join(current_tokens),
                    "label": current_class
                })
                current_tokens = []
                current_class = None
            continue

        tag_type, label = tag.split("-", 1)

        if tag_type == "B":  # start new
            if current_tokens:
                entities.append({
                    "entity": " ".join(current_tokens),
                    "label": current_class
                })
            current_tokens = [tok]
            current_class = label

        elif tag_type == "I":  # add 
            if current_class == label: 
                current_tokens.append(tok)
            else:
                if current_tokens:
                    entities.append({
                        "entity": " ".join(current_tokens),
                        "label": current_class
                    })
                current_tokens = [tok]
                current_class = label

    if current_tokens:
        entities.append({
            "entity": " ".join(current_tokens),
            "label": current_class
        })

    return entities
```

```python
sample_df['entities'] = sample_df.apply(
    lambda row: bio_to_entities(row['sentence_text'], row['ner_tags_str']),
    axis=1
)

sample_df.head()
```
