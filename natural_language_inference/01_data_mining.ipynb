```python
import pandas as pd
```

```python
df = pd.read_csv("../data/stanford-natural-language-inference-corpus/snli_test.csv")
df = df[["sentence1", "sentence2", "gold_label"]]
```

```python
display(df.head())
display(df['gold_label'].value_counts())
display(df['gold_label'].value_counts(normalize=True))
display(df.info())
```

# prepare data for llm labeling

```python
df_sample_1500 = df.sample(n=1500, random_state=42)
df_sample_50 = df.sample(n=50, random_state=42)
```

```python
display(df_sample_1500['gold_label'].value_counts())
display(df_sample_50['gold_label'].value_counts())
```

```python
df_sample_1500.to_csv("../data/stanford-natural-language-inference-corpus/snli_test_sample_1500.csv", index=False)
df_sample_50.to_csv("../data/stanford-natural-language-inference-corpus/snli_test_sample_50.csv", index=False)
```
