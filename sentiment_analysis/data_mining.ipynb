# data mining

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

upload dataset amdb 

```python
train_pos_data = pd.read_csv('../data/amdb_dataset/train/train_pos_data.csv')
train_neg_data = pd.read_csv('../data/amdb_dataset/train/train_neg_data.csv')

train_pos_data.head()
```

```python
data = pd.concat([train_pos_data, train_neg_data], ignore_index=True)
data = data.sample(frac=1).reset_index(drop=True)  # mox data
data.head()
```

```python
categories = range(1, 11)
plt.figure(figsize=(10, 6))

sns.countplot(x='mark', data=data, order=categories)
plt.title('Distribution of Reviews by Category')
plt.xlabel('Marks')
plt.ylabel('Number of Reviews')
plt.show()
```
